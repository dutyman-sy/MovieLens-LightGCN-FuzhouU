# Quick Start - 3步跑通实验

## 第1步：安装依赖
```bash
pip install torch pandas numpy matplotlib scipy
```

## 第2步：下载数据
- 下载地址：https://grouplens.org/datasets/movielens/100k/
- 操作：点击“ml-100k.zip”下载，解压后文件夹命名为 `ml-100k`，和本代码放在同一目录

## 第3步：运行实验
1. 先执行数据预处理：
```bash
python data_preprocess.py
```
2. 再执行模型训练+测试：
```bash
python model_core.py
```
